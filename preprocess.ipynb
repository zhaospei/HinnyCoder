```python
import pandas as pd

def get_id(dir_path='scg-data', type='random'):
    with open(f'{dir_path}/{type}/train_id.txt') as file:
        train_id = [line.rstrip() for line in file]
    with open(f'{dir_path}/{type}/test_id.txt') as file:
        test_id = [line.rstrip() for line in file]
    return train_id, test_id
df = pd.read_parquet(f'scg-data/scg-data-processed.parquet', engine='fastparquet')
train_id, test_id = get_id(dir_path='scg-data', type='random')
```

```python
from tqdm import tqdm
import re

def write_string_to_file(absolute_filename, string):
    with open(absolute_filename, 'w') as fout:
        fout.write(string)

examples = []

for index in tqdm(test_id):
    df_commit = df[df['index']==index]
    source_seq = ''
    target_seq = ''
    for _, row in df_commit.iterrows():
        source_seq = ' '.join(row['nl'].split()[:29])
        target_seq = row['code']
    examples.append({'func_code': target_seq, 'prompt': source_seq + ' <gen> '})

import json
def dump_to_file(obj, file):
    with open(file,'w+') as f:
        for el in obj:
            f.write(json.dumps(el)+'\n')
            
dump_to_file(examples,'test.jsonl')
```
